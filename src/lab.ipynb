```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

image_path = 'maze.jpg'

# Load the image
image = cv2.imread(image_path)
# Display the image
plt.figure(1)
plt.imshow(image)



plt.show()



# Transform

pts1 = np.float32([[100,200],[2300,200],[100,1400],[2300,1400]]) # four points on the first image
pts2 = np.float32([[0,0],[2400,0],[0,1580],[2400,1580]]) # four points on the second image

H = cv2.getPerspectiveTransform(pts1,pts2) # homography matrix
image_transformed = cv2.warpPerspective(image, H, (2400,1580))


plt.figure(2)
plt.imshow(image_transformed)
plt.show()





# HSV
image_hsv = cv2.cvtColor(image_transformed, cv2.COLOR_BGR2HSV)
plt.figure(3)
plt.imshow(image_hsv)
plt.show()



# Wall Detection

# Colour HSV Boundaries
wall_lower = np.array([0, 0, 0])
wall_upper = np.array([80, 80, 90])
wall_mask = cv2.inRange(image_transformed, wall_lower, wall_upper)

image_walls = cv2.bitwise_and(image_hsv, image_hsv, mask=wall_mask)

# print(image_walls[300:350, 100])

segment = image_walls[300:350, 100]


plt.figure(5)
plt.imshow(image_walls)
plt.xticks(np.arange(0, 2400, 100))
plt.yticks(np.arange(0, 1600, 100))
plt.show()


output = [" --- --- --- --- --- --- --- --- --- ", 
"|", 
" ", 
"|", 
" ", 
"|", 
" ", 
"|", 
" ", 
"|", 
" --- --- --- --- --- --- --- --- --- "]

rowWidth = 2400
colWidth = 1600
buffer = 35

def checkWallRight(rowIndex, colIndex):
	wallPointX = rowWidth/9 * (colIndex + 1)
	wallPointY = colWidth/5 * rowIndex + colWidth/10
	# print(f"WallPointXY: ({wallPointX} , {wallPointY})")

	val = image_walls[int(wallPointY), (int(wallPointX) - buffer):(int(wallPointX) + buffer)]
	
	# if rowIndex == 0 and colIndex == 2:
	# 	print(wallPointX)
	# 	print(wallPointY)
	

	if np.size(np.where(val != 0)) != 0:
		return True

	return False

def checkWallDown(rowIndex, colIndex):
	wallPointX = rowWidth/9 * colIndex + rowWidth/18
	wallPointY = colWidth/5 * (rowIndex + 1)

	val = image_walls[(int(wallPointY) - buffer):(int(wallPointY) + buffer), int(wallPointX)]


	
	if np.size(np.where(val != 0)) != 0:
		return True



	return False
    

for rowIndex in range(0, 5):
	for colIndex in range(0, 9):
		# Check wall right, if not left most col
		outputRow = rowIndex * 2 + 1
                
		# print(f"({rowIndex} , {colIndex}) Right Wall = {checkWallRight(rowIndex, colIndex)}  Down Wall = {checkWallDown(rowIndex, colIndex)}")

		if (checkWallRight(rowIndex, colIndex) or colIndex == 8):
			output[outputRow] += "   |"
		else:
			output[outputRow] += "    "


		# Check wall down, if not last row
		if (rowIndex == 4):
			continue
		elif (checkWallDown(rowIndex, colIndex)):
			output[outputRow + 1] += "--- "
		else:
			output[outputRow + 1] += "    "

for line in output:
    print(line)


```
